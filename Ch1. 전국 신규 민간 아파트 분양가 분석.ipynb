### 연도와 월을 분리하기

+ pandas의 string-handling 사용하기

```python
date = "2013년12월"
date
```

```python
#split을 통해 '년'을 기준으로 분리하기
date.split('년')
```

```python
#리스트의 인덱싱을 사용하여 연도만 가져오기
date.split("년")[0]
```

```python
date.split("년")[-1].replace("월", "") #마지막 가져옴
```

```python
#parse_year 이라는 함수를 만듦.
#연도만 반환하도록 하며, 반환하는 데이터는 int타입이 되도록 함
def parse_year(date):
    year = date.split("년")[0]
    year = int(year)
    return year

y= parse_year(date)
print(type(y))
y
```

```python
#제대로 분리가 되었는지 parse_year 함수를 확인함
parse_year(date)
```

```python
def parse_month(date):
    month = date.split("년")[-1].replace("월", "")
    month = int(month)
    return month

parse_month(date)
```

* apply() : pandas객체에 열 혹은 행에 대해 함수를 적용하게 해주는 메서드

```python
# df_first_melt변수에 담긴 데이터프레임에서
# apply를 활용해 연도만 추출해서 새로운 컬럼에 담기

df_first_melt["연도"] = df_first_melt["기간"].apply(parse_year)
df_first_melt.head(1)
```

```python
# df_first_melt변수에 담긴 데이터프레임에서
# apply를 활용해 월만 추출해서 새로운 컬럼에 담기

df_first_melt["월"] = df_first_melt["기간"].apply(parse_month)
df_first_melt.head(1)
```

```python
# 최근 데이터가 담긴 df_last에는 전용면적이 있음
# 이전 데이터에는 전용면적이 없기 때문에 "전체"만 사용하도록 함
# loc를 사용해서 전체에 해당하는 면적만 copy해서 df_last_prepare 변수에 담기
df_last_prepare = df_last.loc[df_last["전용면적"] == "전체" , cols].copy()
df_last_prepare.head(1)
#loc 사용하면 행과 열로 가져올 수 있음
```

* copy() : 명시적으로 복사해줌 / df_last_prepare바껴도 df_last 변경 안됨

```python
#df_first_melt에서 공통된 컬럼만 가져온 뒤
# copy로 복사해서 df_first_prepare 변수에 담기
df_first_prepare = df_first_melt[cols].copy()
df_first_prepare.head(1)
```

### concat으로 데이터 합치기

```python
#df_first_prepare 와 df_last_prepare 합치기
df = pd.concat([df_first_prepare, df_last_prepare])
df.head()
```

```python
#제대로 합쳐졌는지 확인
df.tail()
df.head()
```

```python
#연도별로 데이터가 몇개씩 있는지 세기
df["연도"].value_counts(sort=False)
# ==> 2015년에 데이터 몇개 누락되어있음
```

### Pivot_table 사용하기

```python
#연도를 인덱스로, 지역명을 컬럼으로 평당분양가격을 피봇테이블로 그려보기
t = pd.pivot_table(df, index="연도", columns ="지역명", values = "평당분양가격").round()
t
```

```python
#위에서 그린 피봇테이블을 히트맵으로 표현해 봅니다.
plt.figure(figsize = (15,7))
sns.heatmap(t, cmap="Blues", annot = True, fmt = ".0f")
#cmap() : 한가지색으로만 표현하기
# fmt = ".0f" : float형태로 표현하되 소수점 없이 표현해라
```

```python
#transpose 를 사용하면 행과 열을 바꿀 수 있음
t.T
# 혹은 t.transpose()
```

```python
#바뀐 행과 열을 히트맵으로 표현함
plt.figure(figsize = (15,7))
sns.heatmap(t.T, cmap="Blues", annot = True, fmt = ".0f")
```

* ==> 서울이 해가 갈수록 평균 분양가가 올라가고 있는 것을 확인할 수 있음

```python
#Groupby로 그리기. 인덱스에 ["연도", "지역명"]을 넣고 그리기
g = df.groupby(["연도", "지역명"])["평당분양가격"].mean().unstack().round()
g
```

* unstack() : 맨 마지막 인덱스 값을 컬럼으로 만들어줌

```python
plt.figure(figsize = (15,7))
sns.heatmap(g.T, annot=True, fmt = ".0f", cmap="Greens")
```
